```python
from bs4 import BeautifulSoup # for pulling data out of html and xml pages
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage=requests.get("https://www.flipkart.com/camera-accessories/tripods/pr?sid=jek,6l2,ce6&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_01b7bb5b-8b7d-47b9-a0bb-ac08a949c68a_2_372UD5BXDFYS_MC.XRI8PU4LB3H7&otracker=hp_rich_navigation_7_2.navigationCard.RICH_NAVIGATION_Electronics~Cameras%2B%2526%2BAccessories~Camera%2Btripods_XRI8PU4LB3H7&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_7_L2_view-all&cid=XRI8PU4LB3H7")
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify())
#prettify:how html tags are arranged is displayed
```

```python
names=soup.find_all('a', class_="s1Q9rs")
```

```python
names
```

```python
tripod=[]
for i in range(0,len(names)):
  tripod.append(names[i].get_text())
```

```python
tripod
```

```python
specification=soup.find_all('div', class_="_3Djpdu")
```

```python
specification
```

```python
specifications=[]
for i in range(0,len(specification)):
  specifications.append(specification[i].get_text())
```

```python
specifications
```

```python
rating=soup.find_all('a', class_="_8VNy32")
```

```python
rating
```

```python
ratings=[]
for i in range(0,len(rating)):
  ratings.append(rating[i].get_text())
```

```python
ratings
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['Tripod names']=tripod
```

```python
df
```

```python
df['Specifications']=specifications
```

```python
df
```

```python
df['Price & Offer']=ratings
```
