```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
iris.head()
```

pairgrid
it is subplot grid for plotting pairwise relationships in dataset

```python
sns.PairGrid(iris)
```

```python
map this to grid
```

```python
y=sns.PairGrid(iris)
y.map(plt.scatter)
```

map to upper,lower,and diagonal in histogram



```python
y=sns.PairGrid(iris)
y.map_upper(plt.scatter)
y.map_diag(plt.hist)
y.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

```python
sns.pairplot(iris,hue='species',palette='coolwarm')
```

```python
tip_dataset=sns.load_dataset('tips')
tip_dataset.head()
```

```python
m=sns.FacetGrid(tip_dataset,col="time",row="smoker")
m=m.map(plt.hist,"day")
```
