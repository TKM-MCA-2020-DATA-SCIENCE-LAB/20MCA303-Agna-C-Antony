```python
import numpy as np #importing numpy library
x=np.linspace(0,5,11)
y =x ** 2
```

```python
import matplotlib.pyplot as p #importing matplotlib 
p.plot(x,y,'g')
p.xlabel("x axis title here")
p.ylabel("y axis title here")
p.title(" x plotted against y")
p.show()
```

```python
p.subplot(1,1,1)
p.plot(x,y,'y*-')
p.subplot(2,4,2)
p.plot(x,y,'r--')
```

matplot using object oriented concept



```python
#create figure(empty canvas)
fig1=p.figure()
```

```python
#to add axes
axes=fig1.add_axes([0.2,0.2,0.8,0.8])
```

```python
f=p.figure()
axes=f.add_axes([0.5,0.5,0.8,0.8])
axes.plot(x,y,'y')
axes.set_xlabel('speed')
axes.set_ylabel('time')
axes.set_title('speed vs time')
```

```python
f=p.figure()
axes1=f.add_axes([0.1,0.1,0.8,0.8])#large
axes2=f.add_axes([0.2,0.6,0.2,0.2])#inner
axes1.plot(x,y,'y')
axes1.set_xlabel('speed')
axes1.set_ylabel('time')
axes1.set_title('speed vs time')
axes2.plot(y,x,'r--')
axes2.set_xlabel('s')
axes2.set_ylabel('t')
axes2.set_title('speed vs time')
```

To create subplots()


```python
f,axes=p.subplots()
axes.plot(x,y,'b*--')
axes.set_xlabel('s')
axes.set_ylabel('t')
axes.set_title('s vs t')
```

```python
f,axes=p.subplots()
axes.plot(x,y,'b*--')
axes.set_xlabel('s')
axes.set_ylabel('t')
axes.set_title('s vs t')
```

```python
axes
```

```python
f,axes=p.subplots(1,3)
for axe in axes:
    axe.plot(y,x,'r*--')
    axe.set_xlabel('s')
    axe.set_ylabel('t')
    axe.set_title('s vs t')
    f.tight_layout()
```

```python
f,axes=p.subplots(1,4,figsize=(10,2),dpi=100)

f.tight_layout()
```

```python
f,axes=p.subplots(1,3,figsize=(10,2),dpi=100)
for axe in axes:
    axe.plot(x,y,'r*--')
    axe.set_xlabel('s')
    axe.set_ylabel('t')
    axe.set_title('s vs t')
    f.tight_layout()
```

```python
fig=p.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=1)
```
