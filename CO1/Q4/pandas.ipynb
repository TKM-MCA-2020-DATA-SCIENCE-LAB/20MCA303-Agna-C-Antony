To convert list,numpy array,dictionary into series.

```python
import numpy as n
import pandas as p
l1=['x','y','z']
array=n.array([12,15,18])
dict={1:'agna',2:'nidhi',3:'sree'}
```

```python
p.Series(data=l1)
```

```python
p.Series(data=l1,index=l1)
```

```python
p.Series(array,index=l1)
```

```python
p.Series(dict)
```

```python
s1=p.Series([4,5,6],index=['Thrissur','kollam','kozikode'])
s2=p.Series([5,34,5,6],index=['Thrissur','kochi','kollam','kozikode'])
s1+s2
```
