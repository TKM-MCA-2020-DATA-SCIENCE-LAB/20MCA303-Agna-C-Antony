```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as s
```

```python
dataset=pd.read_csv('/content/Heart_Disease_Prediction.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(dataset.drop('Heart Disease',axis=1))
```

```python
x=scalar.transform(dataset.drop('Heart Disease',axis=1))
```

```python
x
```

```python
newdataset=pd.DataFrame(x,columns=dataset.columns[:-1])
```

```python
newdataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=newdataset
y=dataset['Heart Disease']
```

```python
x
```

```python
y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=20)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,CategoricalNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
```

```python
scores=[]
classifier=GaussianNB()
classifier.fit(X_train,Y_train)
y_pred=classifier.predict(X_test)
scores.append(accuracy_score(Y_test,y_pred))
cm=confusion_matrix(Y_test,y_pred)
cm
```

```python
print(classification_report(Y_test,y_pred))
```
